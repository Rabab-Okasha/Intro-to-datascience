```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Draw a line in a diagram from position (0,0) to position (6,250):

```python
x = np.array([0, 6])
y = np.array([0, 250])
plt.plot(x, y)
plt.show()
```

## Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
plt.plot(x, y)
plt.show()
```

## Plotting without x-points:

```python
y = np.array([3, 8, 1, 10, 5, 7])
plt.plot(y, color = 'r')
plt.show()
# scale of x changes, complier puts on default values
```

## Plot with a 20.5pt wide line:

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
plt.plot(x, y, linewidth = 20.5, linestyle = '-.', color = 'green')
plt.show()
```

## Draw two lines by specifying the x- and y-point values for both lines:

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

## Draw a bar plot consisting of 4 bars:

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.bar(x, y)
plt.show()
```

## Change the color of the bars to red, make them horizontal with height=0.2:

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, width = 0.2, color = 'red')
plt.show()
```

## Create random arrays with 100 values for x-points, y-points, colors and sizes:

```python
from numpy import random
x = np.random.randint(0, 100, size = 100)
y = np.random.randint(0, 100, size = 100)
color = np.random.randint(0, 100, size = 100)
sizes = np.random.randint(0, 100, size = 100)

plt.scatter(x, y, c = color, s = sizes, alpha = 0.75, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

## Bubble plot

```python
price = [600, 250, 45, 120, 80, 450]
rating = [4.7, 4.5, 4.8, 4.3, 3.8, 4.6]
units_sold = [800, 4500, 15000, 2500, 3000, 1200]

bubble_size = [ p * 0.1 for p in units_sold]
plt.scatter(price, rating, s = bubble_size, alpha = 0.5, c = 'r')
plt.show()
```

## Box plot

```python
data = np.random.normal(100, 25, 200)
np.random.seed(5)
figure = plt.figure(figsize = (10, 8))
plt.boxplot(data, vert = False, patch_artist = True, notch = True)
plt.show()
```

## Create multiple box plots simultaneously in the same file

```python
data1 = np.random.normal(100, 25, 200)
data2 = np.random.normal(100, 20, 100)
data3 = np.random.normal(110, 10, 220)
data4 = np.random.normal(190, 25, 250)
dataset = [data1, data2, data3, data4]
figure = plt.figure(figsize = (8, 8))
ax = figure.add_axes([0, 0, 1, 1])
ax.boxplot(dataset, vert = False, patch_artist = True, notch = True)
plt.show()
```

```python
from pywaffle import Waffle
```

```python
df = pd.DataFrame({
    'Country' : ['Argentina','Brazil', 'Cuba' ,'Peru'],
    'Number' : [21, 300, 1200, 500]
}) 
```

```python
fig = plt.figure(
    FigureClass = Waffle,
    rows = 10,
    values = list(df.Number / 5),
    labels = list(df['Country']),
    figsize = (12, 8),
    legend = {'bbox_to_anchor' : (1, 1)}
    # legend = {'loc' : 'upper left', 'bbox_to_anchor': (1, 1)}
)
plt.show()
```

```python
df = pd.read_csv('Files/Restaurant_Reviews - Restaurant_Reviews.tsv', sep = '\t')
df.head()
```

```python
text = " ".join(cat for cat in df['Review'])
text
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_cloud = WordCloud(
    width = 3000,
    height = 2000,
    random_state = 1,
    background_color = 'white',
    colormap = 'viridis',
    collocations = False,
    stopwords = STOPWORDS,
).generate(text)
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
