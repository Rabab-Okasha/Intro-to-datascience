```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset.head()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x = 'total_bill', y = 'tip', data = dataset)
```

```python
# regression plot
sns.lmplot(x = 'total_bill', y = 'tip', col = 'sex', row = 'time', hue = 'smoker', data = dataset)
plt.show()
```

```python
Iris = sns.load_dataset('iris')
Iris.head()
```

```python
Iris.info()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x = 'sepal_width', y = 'sepal_length', hue = 'species', data = Iris)
plt.title('Regression plot for Iris dataset')
plt.xlabel('Sepal Width(cm)')
plt.ylabel('Sepal length(cm)')
plt.show()
```
